```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
# To plot
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
from sklearn.metrics import accuracy_score
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing datasets  
dfx= pd.read_csv('stocknet_trn_data (1).csv')  
dfx.head()
```

```python
dfy=pd.read_csv('stocknet_trn_data_targets (1).csv',header=None)
dfy = dfy.rename(columns ={1:"Close"})
dfy.head()
```

```python
dfx=dfx.dropna()
dfy=dfy.dropna()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaled = MinMaxScaler().fit_transform(dfx)
dfx = pd.DataFrame(scaled,columns = dfx.columns)
dfx
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfx,dfy['Close'], test_size= .30, random_state=1)
```

```python
# Create and train the Support Vector Machine (Regressor) 
svr_rbf = SVR(kernel='poly', C=1, epsilon=10 ) 
svr_rbf.fit(X_train, y_train)
```

```python
# Testing Model: Score returns the coefficient of determination R^2 of the prediction. 
# The best possible score is 1.0
svm_confidence = svr_rbf.score(X_test, y_test)
print("svm confidence: ", svm_confidence)
```
